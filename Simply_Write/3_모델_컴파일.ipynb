## **컴파일**이란 특정 프로그래밍 언어로 쓰여 있는 문서를 다른 프로그래밍 언어로 옮기는 언어 번역 프로그램이다.

```python
# 라이브러리 불러오기
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
# 모델 생성 (다중 분류)
model = Sequential([
                    Flatten(input_shape = (28, 28)),
                    Dense(64, activation = "relu"), # "units = " 생략가능
                    Dense(10, activation = "softmax") 
])
```

```python
model.summary()
```

### - Loss: 손실함수 (오차 측정 방법.)
### - Optimizer: 최적화 (훈련하는 동안 Loss값을 최소화 시켜야하기위해 사용.)
### - Metrics: 지표

```python
# 모델 컴파일 방법 1
model.compile(optimizer = 'adam', # "옵티마이저로 adam을 사용" 
              loss = "sparse_categorical_crossentropy", # loss: Error 값. 즉, 낮을수록 결과값이 좋다.
              metrics = ["accuracy"]) # 정확도
```

```python
# 모델 컴파일 방법 2
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics = tf.keras.metrics.Accuracy())
```

## Loss(손실함수):
### 훈련(학습)하는 동안 오차를 측정한다. 최소화 해야한다.

## " sparse_categorical_crossentropy "  VS  " categorical_crossentropy "
### ---------------------------------------------------------------------------------------------------------------------------
### sparse_categorical_crossentropy:
### -> Label: 1, 2, 3
### ------------------------------------------------------------------
### categorical_crossentropy:
### -> Label: [1,0,0], [0, 1, 0], [0, 0, 1]

```python
# 다중분류
model.compile(optimizer = 'adam', # "옵티마이저로 adam을 사용" 
              loss = "categorical_crossentropy", # loss: Error 값. 즉, 낮을수록 결과값이 좋다.
              metrics = ["accuracy"]) # 정확도
```

```python
# 이진분류
model.compile(optimizer = 'adam', # "옵티마이저로 adam을 사용" 
              loss = "binary_crossentropy", # loss: Error 값. 즉, 낮을수록 결과값이 좋다.
              metrics = ["accuracy"]) # 정확도
```

```python
# 회귀
model.compile(optimizer = 'adam', # "옵티마이저로 adam을 사용" 
              loss = "mse", # loss: Error 값. 즉, 낮을수록 결과값이 좋다. / "mse" or "mae"
              metrics = ["mae"]) # 정확도
```
